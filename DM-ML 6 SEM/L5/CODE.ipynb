```python
xmean=df.mean()['Outdoor_temp_x']
ymean=df.mean()['Supply_temp_y']
df.mean()
```

```python
df['y-ymean']=df['Supply_temp_y']-ymean
df['x-xmean']=df['Outdoor_temp_x']-xmean
df['(y-ymean)^2']=df['y-ymean']**2
df['(x-xmean)^2']=df['x-xmean']**2
df['(y-ymean)(x-xmean)']=df['y-ymean']*df['x-xmean']
df
```

```python
df.plot.scatter(x='Outdoor_temp_x', y='Supply_temp_y', c='No.', colormap='rainbow')
```

```python
x=df.drop(['No.'], axis=1)
y=df['Supply_temp_y']
model=LinearRegression()
model.fit(x, y)
y_pred = model.predict(x) #predicted values
```

```python
df['y-ymean']=df['Supply_temp_y']-ymean
df['x-xmean']=df['Outdoor_temp_x']-xmean
df['(y-ymean)^2']=df['y-ymean']**2
df['(x-xmean)^2']=df['x-xmean']**2
df['(y-ymean)(x-xmean)']=df['y-ymean']*df['x-xmean']
df['y_pred']=y_pred
df
```
